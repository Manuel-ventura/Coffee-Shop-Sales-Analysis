# Maven Roasters Sales Analysis

## Introduction
This project aims to analyze the sales data of Maven Roasters to answer the following key questions:

- ***How have Maven Roasters sales trended over time?***
- ***Which days of the week tend to be busiest, and why do you think that's the case?***
- ***Which products are sold most and least often? Which drive the most revenue for the business?***

By addressing these questions, we can gain insights into sales patterns, peak business days, and product performance, ultimately helping Maven Roasters make informed business decisions.

## Objectives

- Sales Trends Over Time: Identify the overall sales trend, including any seasonal patterns or growth trends.
- Busiest Days of the Week: Determine which days of the week have the highest sales and analyze possible reasons for this trend.
- Product Performance: Identify the best and worst-selling products and analyze which products contribute most to the revenue.

## Conclusion
By the end of this analysis, we will have a clearer understanding of Maven Roasters' sales dynamics, enabling more informed business decisions. We will know the historical sales trends, the busiest days, and the products that significantly impact revenue.

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the data
df = pd.read_excel(r'Coffee Shop Sales.xlsx')
df.head()
```

```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
df.info()
```

```python
df['transaction_time'] = df['transaction_time'].astype(str)
df['transaction_time'] = pd.to_timedelta(df['transaction_time'])
```

```python
df.info()
```

```python
df['sales'] = df['transaction_qty'] * df['unit_price']
```

```python
df['datetime'] = df['transaction_date'] + df['transaction_time']
df.head()
```

```python
df.describe()
```

```python
df_location = df.groupby('store_location').agg({
    'sales': 'sum',
    'transaction_id': 'count'
})

df_location
```

```python
daily_sales_by_location = df.groupby(['transaction_date', 'store_location'])['sales'].sum().unstack()
daily_sales_by_location
```

```python
# Plot the daily sales by store location
daily_sales_by_location.plot(figsize=(14, 8), title='Dayli Sales by Store Location')

plt.xlabel('Date')
plt.ylabel('Total Sales ($)')
plt.legend(title='Store Location')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

#Show the plot
plt.show()
```

```python
plt.figure(figsize=(14, 8))

plt.stackplot(daily_sales_by_location.index, daily_sales_by_location.T, labels=daily_sales_by_location.columns)

plt.title('Dayli Sales by Store Location (Staked Area Chart)')
plt.xlabel('Date')
plt.ylabel('Total Sales ($)')
plt.legend(title='Store Location', loc='upper left')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Calculate daily sales
# Group the dataframe by 'transaction_date' and sum the 'sales' for each date
# Reset the index to convert the groupby object back to a dataframe

daily_sales = df.groupby('transaction_date')['sales'].sum().reset_index()

df['day_of_week'] = df['datetime'].dt.day_name()

# Calculate weekly sales
weekly_sales = df.groupby('day_of_week')['sales'].sum().reindex(
        ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']).reset_index()

# Calculate monthly sales
df['month'] = df['datetime'].dt.to_period('M')
monthly_sales = df.groupby('month')['sales'].sum().reset_index()


# Display the calculated daily, weekly, and monthly sales dataframes
daily_sales, weekly_sales, monthly_sales
```

```python
# Monthly Sales
monthly_sales.plot(figsize=(14, 8), title='Monthly Sales')

plt.xlabel('Date')
plt.ylabel('Total Sales ($)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Weekly Sales
weekly_sales.plot(figsize=(14, 8), title='Weekly Sales')

plt.xlabel('Date')
plt.ylabel('Total Sales ($)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Dayli Sales
daily_sales.plot(figsize=(14, 8), title='Daily Sales')

plt.xlabel('Date')
plt.ylabel('Total Sales ($)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Calculate weekly sales by store location
weekly_sales = df.groupby(['day_of_week', 'store_location'])['sales'].sum().unstack()

days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekly_sales = weekly_sales.reindex(days_order)
weekly_sales
```

```python
weekly_sales[["Astoria", "Hell's Kitchen", "Lower Manhattan"]].plot(kind='bar',figsize=(14, 8))

plt.xlabel('Day of the Week', fontweight='bold')
plt.ylabel('Total Sales ($)', fontweight='bold')
plt.title('Weekly Sales by Store Location')
plt.legend(title='Store Location', loc='upper left')

#show the plot
plt.tight_layout()
plt.show()
```

```python
df_category = df.groupby('product_category').agg({
    'sales': 'sum',
    'transaction_id': 'count'
}).sort_values('sales', ascending = False)

df_category
```

```python
# Extract the hour from the 'datetime' column and create a new column 'hour'
df['hour'] = df['datetime'].dt.hour

# Calculate hourly sales by product category
hourly_sales_by_product = df.groupby(['hour', 'product_category'])['sales'].sum().unstack()

hourly_sales_by_product.plot(kind='bar', stacked=True, figsize=(14, 8), colormap='viridis')

plt.title('Hourly Sales by Product Category')
plt.xlabel('Hour of the Day')
plt.ylabel('Total Sales ($)')
plt.legend(title='Product Category', loc='upper right')
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
df_category['avg_sales'] = df_category['sales']/df_category['transaction_id']
df_category
```

```python
sales_by_product = df.groupby('product_type').agg({
    'sales': 'sum',
    'unit_price': 'mean'
}).sort_values('sales', ascending = False)

sales_by_product
```

```python
product_sales_df = df.groupby(['product_category', 'product_type'])['sales'].sum().reset_index()
product_sales_df
```

```python
plt.figure(figsize=(14, 8))
sns.barplot(data=product_sales_df, x='product_category', y='sales', hue='product_type')

#Add labels and title
plt.title('Product Sales by Category and Type')
plt.xlabel('Product Category')
plt.ylabel('Total Sales ($)')
plt.legend(title='Product Type', loc='upper right')

#Show the plot
plt.tight_layout()
plt.show()
```

```python
# Stacked Bar Chart
pivot_df = product_sales_df.pivot(index='product_category', columns='product_type', values='sales').fillna(0)


# Define a custom color palette using tab20b and tab20c
colors = plt.get_cmap('tab20b').colors + plt.get_cmap('tab20c').colors

# Select only the first 28 colors
colors = colors[:28]


pivot_df.plot(kind='bar', stacked=True, figsize=(14, 8), color=colors)
plt.title('Stacked Bar Chart of Sales by Product Category and Type')
plt.xlabel('Product Category')
plt.ylabel('Total Sales ($)')
plt.legend(title='Product Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Create the pivoted DataFrame
pivot_df = product_sales_df.pivot(index='product_category', columns='product_type', values='sales').fillna(0)

colors = plt.get_cmap('tab20b').colors + plt.get_cmap('tab20c').colors

num_product_types = len(product_sales_df['product_type'].unique())
colors = colors[:num_product_types]

# Create a dictionary to assign colors to each product type
product_types = product_sales_df['product_type'].unique()
color_dict = dict(zip(product_types, colors))

# Create individual plots for each category using the same color palette
categories = product_sales_df['product_category'].unique()
num_categories = len(categories)
fig, axes = plt.subplots(num_categories, 1, figsize=(10, num_categories * 5))

for ax, category in zip(axes, categories):
    category_df = product_sales_df[product_sales_df['product_category'] == category]
    sns.barplot(data=category_df, x='product_type', y='sales', ax=ax, hue='product_type', dodge=False, palette=color_dict)
    ax.set_title(f'Sales by Product Type in {category}')
    ax.set_xlabel('Product Type')
    ax.set_ylabel('Total Sales ($)')
    ax.tick_params(axis='x')

plt.tight_layout()
plt.show()
```
